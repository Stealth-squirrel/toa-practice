##Упражнение 29
Вы управляете строительством рекламных щитов на дороге с интенсивным
движением, идущей с запада на восток на протяжении M км. Возможные точки для
установки рекламных щитов определяются числами x1, x2, ..., xn, каждое из которых
принадлежит интервалу [0, M] (эти числа определяют позицию в км. от западного
конца дороги). Разместив рекламный щит в точке xi
, вы получите прибыль ri > 0. По
правилам дорожной службы округа два рекламных щита не могут находиться на
расстоянии 5 км. и менее. Требуется разместить рекламные щиты на подмножестве
точек так, чтобы обеспечить максимальную прибыль с учетом ограничения.
Пример. Предположим, M = 20, n = 4,
(x1, x2, x3, x4) = (6, 7, 12, 14), и (r1, r2, r3, r4) = (5, 6, 5, 1).
В оптимальном решении рекламные щиты размещаются в точках x1 и x3, с
суммарной прибылью 10.
Предложите алгоритм, который получает экземпляр этой задачи и возвращает
максимальную суммарную прибыль, которая может быть получена для любого
действительного подмножества точек. Время выполнения алгоритма должно быть
полиномиальным по n.

```python
def max_profit(x, r):
    n = len(x)
    prev = [-1] * n   # индекс последней точки, с которой у i расстояние >5
    j = 0
    for i in range(n):
        while j < i and x[i] - x[j] > 5:
            j += 1
        prev[i] = j - 1

    max_profit = [0] * (n + 1)   # max_profit[i] - максимальная прибыль, которую можно получить, рассматривая первые i точек
    for i in range(1, n + 1):
        best = max_profit[i-1]   # не берём i-ю точку
        take = r[i-1]    # берём i-ю точку
        if prev[i-1] != -1:
            take += max_profit[prev[i-1] + 1]
        best = max(best, take)
        max_profit[i] = best

    return max_profit[n]
```

```python
x = [6, 7, 12, 14]
r = [5, 6, 5, 1]
print(f'Максимальная суммарная прибыль: {max_profit(x, r)}')
```

##Упражнение 30
Имеется ненаправленный граф с n узлами G = (V, E). Напомним, что
подмножество узлов называется независимым множеством, если
никакие два узла не соединяются ребром. В общем виде задача
нахождения большого независимого множества сложна, но при
достаточно «простом» графе она имеет эффективное решение.
Назовем граф G = (V, E) путевым графом, если его узлы могут быть
записаны в виде v1, v2, ..., vn, причем vi и vj соединяются ребром в том
и только в том случае, если i и j различаются ровно на 1. С каждым
узлом vi связывается положительный целый вес wi
.
Целью этого упражнения является решение следующей задачи:
Найдите в путевом графе G независимое множество с максимальным
общим весом.

###Упражнение 30(a)
(a) Приведите пример, показывающий, что
следующий алгоритм не всегда находит независимое
множество с максимальным общим весом.
Жадный алгоритм «приоритетного выбора
максимального веса»

Начать с S, равного пустому множеству

Пока в G остаются узлы

Выбрать узел vi с максимальным весом

Добавить vi в S

Удалить vi и его соседей из G

Конец Пока

Вернуть S

**а) Решение**

Рассмотрим путевой граф из трёх вершин v1−v2−v3​ с весами:

w1=7, w2=15, w3=10


Жадный алгоритм выбирает вершину с максимальным весом – v2 (вес 15). Затем удаляет v2 и её соседей v1,v3.
В итоге получаем множество S={v2} с весом 15.


Оптимальное независимое множество – {v1,v3}, так как они не смежны, их суммарный вес 7+10=17, что больше 15.

Таким образом, жадный алгоритм не всегда находит максимальное по весу независимое множество.

### 30(c)
 Предложите алгоритм, который получает путевой граф G из n узлов с весами и возвращает
независимое множество с максимальным общим
весом. Время выполнения должно быть
полиномиальным по n независимо от значений
весов.

```python
def max_weight_ind_set(weights):
    n = len(weights)
    max_weight = [0] * n   # максимальный суммарный вес независимого множества среди первых i+1 вершин
    max_weight[0] = weights[0]
    max_weight[1] = max(weights[0], weights[1])
    for i in range(2, n):
        max_weight[i] = max(max_weight[i-1], max_weight[i-2] + weights[i])
    return max_weight[-1]
```

```python
weights = [1, 8, 6, 3, 6]
print(f'Максимальный общий вес независимого множества: {max_weight_ind_set(weights)}')
```
